```python
import pandas as pd

df = pd.read_csv('Resources/rotten_tomatoes_movies.csv')
df.head()
```

```python
#split cast members into columns
df1 = df[["cast"]]
df2 = df1.cast.str.split(', ', expand=True)
#Keep first five actors listed
df2 = df2[[0,1,2,3,4]]
df2 = df2.rename(columns={0: "first_actor", 1: "second_actor", 2: "third_actor", 3: "fourth_actor", 4: "fifth_actor"})
df2.head()
```

```python
#split genres
df3 = df[["genre"]]
df4 = df3.genre.str.split(', ', expand=True)
#Keep first five actors listed
df4 = df4[[0,1,2]]
df4 = df4.rename(columns={0: "first_genre", 1: "second_genre", 2: "third_genre"})
df4.head()
```

```python
#drop original cast & genre columns

df = df.drop(['cast','genre'], axis=1)
df.head()
```

```python
#merge actors columns & genre columns

merged = pd.concat([df,df2,df4], axis=1, sort=False)
merged.head()
```
