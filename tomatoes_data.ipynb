```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)

movie_data = "Resources/rotten-tomatoes-movies-and-critics-datasets/rotten_tomatoes_movies.csv"
revenue_data = "Resources/revenue.csv"

# Read the Data
movie_df = pd.read_csv(movie_data)

# Display the data table for preview
movie_df.head(50)
```

```python
# Get just the title and revenue columns
movie_contributors = pd.DataFrame({"title" : movie_df["movie_title"], 
                             "genre" : movie_df["genre"],
                             "cast" : movie_df["cast"],
                             "directors" : movie_df["directors"]})
#movie_contributors


result_list = []
for row_dex, row in movie_df.iterrows():
    actor_list = str(row["cast"])
    #print(actor_list)
    actor_array = actor_list.split(', ')
    #row.cast.str.split(', ', expand=True)
    actor_count = len(actor_array)
    result_list.append({"title" : row["movie_title"],
                       "star_count" : actor_count,
                       "audience_rating" :row["audience_rating"],
                        "audience_count" : row["audience_count"],
                        "audience_top_critics_count" : row["audience_top_critics_count"],
                        "audience_fresh_critics_count" : row["audience_fresh_critics_count"],
                        "audience_rotten_critics_count" : row["audience_rotten_critics_count"]})

movie_crit_data = pd.DataFrame(result_list)
movie_crit_data
```

```python
# Remove rows with revenue less than 10,000
revenue = pd.read_csv(revenue_data)
revenue
```

```python
final_list = []
found =0
for row_dex, row in movie_crit_data.iterrows():
    for row2_dex, row2 in revenue.iterrows():
        if (str(row["title"]) == str(row2["title"])):
            final_list.append({"title" : row["title"],
                       "star_count" : row["star_count"],
                       "audience_rating" :row["audience_rating"],
                        "audience_count" : row["audience_count"],
                        "audience_top_critics_count" : row["audience_top_critics_count"],
                        "audience_fresh_critics_count" : row["audience_fresh_critics_count"],
                        "audience_rotten_critics_count" : row["audience_rotten_critics_count"],
                              "revenue" : row2["revenue"]})
            
movie_crit_revenue_data = pd.DataFrame(final_list)
movie_crit_revenue_data
```

```python
movie_crit_revenue_data.to_csv('Resources/tomato_revenue_crit.csv')
```
