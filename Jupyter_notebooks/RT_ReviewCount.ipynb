```python
import pandas as pd

df = pd.read_csv('../Resources/rotten_tomatoes_reviews.csv')
df
```

```python
df = df.groupby(['rotten_tomatoes_link']).count()
df.reset_index(inplace=True)
```

```python
df1 = df[['rotten_tomatoes_link','critic_top','review_content']]
```

```python
df1.head()
```

```python
RT_revenue = pd.read_csv('../Resources/tomato_revenue_crit.csv')
RT_movie_name = pd.read_csv('../Resources/rotten_tomatoes_movies.csv')
```

```python
RT_movie_name = RT_movie_name[['rotten_tomatoes_link','movie_title']]

RT_revenue.set_index('Unnamed: 0', inplace=True)
RT_revenue.head()
```

```python
merged = pd.merge(df1, RT_movie_name, how='left', right_on='rotten_tomatoes_link', left_on='rotten_tomatoes_link')
```

```python
merged.head()
```

```python
##merged.to_csv('../Resources/critic_review_count.csv')
```

```python
revenue_reviews = RT_revenue.merge(merged, how='left', right_on='movie_title', left_on='title')
```

```python
revenue_reviews.drop(columns=['movie_title'], inplace=True)
```

```python
revenue_reviews.rename(columns={'critic_top': 'top_critic_review_count', 'review_content': 'review_count'}, inplace=True)
revenue_reviews = revenue_reviews[['title', 'rotten_tomatoes_link', 'star_count', 'audience_rating', 'audience_count', 'audience_top_critics_count', 'audience_fresh_critics_count', 'audience_rotten_critics_count', 'revenue', 'top_critic_review_count', 'review_count']]
```

```python
revenue_reviews.head()
```

```python
revenue_reviews.to_csv('../Resources/tomato_revenue_crit_reviews.csv')
```
