```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
movie_master_data = "../Resources/rotten-tomatoes-movies-and-critics-datasets/rotten_tomatoes_movies.csv"
movie_data = "../Resources/movie_metadata_adjusted_revenue.csv"
actor_data = "../Resources/actors.csv"

# Read the Data
movie_df = pd.read_csv(movie_data)
actor_df = pd.read_csv(actor_data)
movie_actor_df = pd.read_csv(movie_master_data)

# Display the data table for preview
movie_df.head(5)
#title
```

```python
actor_df.head(5) #actor
```

```python

movie_actor_df.drop_duplicates(inplace=True) #cast
#movie_actor_df.head(5)
```

```python
final_list = []
found =0
for row_dex, row in movie_df.iterrows():
    accumulator = 0
    movie_cast = movie_actor_df.loc[movie_actor_df['movie_title'] == row['title'], ['cast']]
    movie_list = movie_cast['cast']
    for c_line in movie_list:
        cast_list = c_line.split(',')
        for person in cast_list:
            for row_dex2, row2 in actor_df.iterrows():
                if (row2['actor'] == person):
                    accumulator += 1

    final_list.append({"title" : row["title"],
                       "budget" : row["budget"],
                       "popularity" :row["popularity"],
                        "release_date" : row["release_date"],
                        "revenue" : row["revenue"],
                        "runtime" : row["runtime"],
                        "vote_average" : row["vote_average"],
                              "vote_count" : row["vote_count"],
                          "star_power" : accumulator})
    print(accumulator)
    #cast_list = movie_cast.split(',')

result_df = pd.DataFrame(final_list)
```
