```python
import pandas as pd
```

```python
main = pd.read_csv('../Resources/movie_metadata_adjusted_revenue.csv')
#genres = pd.read_csv('../Resources/onehotgenres.csv')
genres = pd.read_csv('../Resources/genres.csv')
critic_reviews = pd.read_csv('../Resources/tomato_revenue_crit_reviews.csv')
```

```python
main
```

```python
critic_reviews = critic_reviews[['title','top_critic_review_count','review_count']]
```

```python
merged = pd.merge(main, genres, left_on='title', right_on='movie_title', how='inner')
merged.drop(columns=['Unnamed: 0_x','Unnamed: 0_y'], axis=1, inplace=True)
list(merged)
```

```python
merged2 = pd.merge(merged, critic_reviews, on='title', how='left')
```
