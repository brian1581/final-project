```python
# Dependency
import pandas as pd
```

```python
# read in CSV
mv_df = pd.read_csv('Resources/the-movies-dataset/movies_metadata.csv')
```

```python
# remove unnecessary columns and rows with null data
mv_dropped = mv_df.drop(columns={'belongs_to_collection', 'genres', 'spoken_languages', 'imdb_id',\
                                 'id', 'homepage', 'production_companies', 'production_countries'\
                                , 'poster_path', 'status', 'original_title', 'video', 'overview'\
                                ,'tagline', 'adult'})
mv_dropped = mv_dropped.dropna()
```

```python
# Convert columns to integers for filtering
mv_dropped['runtime'] = mv_dropped['runtime'].astype(int)
mv_dropped['revenue'] = mv_dropped['revenue'].astype(int)
mv_dropped['vote_count'] = mv_dropped['vote_count'].astype(int)
mv_dropped['budget'] = mv_dropped['budget'].astype(int)
```

```python
# filter via loc
mv_filtered = mv_dropped.loc[(mv_dropped['revenue'] > 0)]
mv_filtered = mv_filtered.loc[(mv_filtered['vote_count'] > 1000)]
mv_filtered = mv_filtered.loc[(mv_filtered['budget'] > 0)]
mv_filtered = mv_filtered.loc[(mv_filtered['original_language'] == 'en')]
```

```python
# extract the year only from the release date
mv_filtered['release_date'] = pd.to_datetime(mv_filtered['release_date'])
mv_filtered['release_date'] = mv_filtered['release_date'].map(lambda x: x.year)
```

```python
# set title as the index, sort by vote scores and remove the language column now that we have filtered by language
df_final = mv_filtered.sort_values(by=['vote_average'], ascending=False)
df_final = df_final.drop(columns={'original_language'})
df_final = df_final.set_index('title')
df_final
```

```python
# Export to CSV
df_export = df_final.to_csv(r'Resources/final_movie_metadata.csv', index = True, header=True)
```
